## **1.3 Training & Testing Split**

Splitting a Machine Learning algorithm into a train-test split allows one to evaluate the performance of the algorithm. It is a technique commonly used when faced with classification or regression problems. 

The technique involves dividing a dataset into 2 subsets, the training dataset and the test dataset. The training dataset is the sample of data that is used to fit the model. This dataset will train the model, the model then sees this data and learns from it. The test dataset is the sample of data that is then used to evaluate the fit machine learning model. The test dataset is only used once a Machine Learning model is fully trained. 

Ultimately the objective of using the train-test split will be to gauge how the Machine Learning model performs on new data, i.e., data that was not used when training the model. This technique is suitable when there is a relatively large dataset present, such as the synthetic data we have created, as well as the iris and wine datasets. 


## **1.4 Method**

Within each dataset we are looking to asign new data to their respective classes whereby the dependent variable $y_{i}$ is categorical. Since $y ∈ {1,...,C}$ where $C > 2$ in all three of our datasets, multiclass classification is used (Synthetic: $C = 4$, Iris: $C = 3$, Wine: $C = 3$). In this case, each class is treated as its own binary classification problem whereby the one vs all approach is implemented (Sucky, 2020). For example, in the Iris dataset, if the value in the test set is similar to those values in training set classed as ‘Setosa’, this class will be denoted by 1 while versicolor and verginca will be 0. This process can be explained in more detail mathematically (Taboga, 2021).

$P(y_{i} = 1|x_{i}) = S(x_{i}\beta) $

Whereby the conditional probability of the dependent variable $y_{i}$ is equal to 1 given the feature values of $x_{i}$. This is equal to the sigmoid function $S$ times the linear combination of inputs of the feature values $x_{i}$ and the beta coefficient $\beta$. In this case, observations would be assigned to class 1.

Similarly, the observation would be assigned to class 0 when:

$P(y_{i} = 0|x_{i}) = 1 - P(y_{i} = 1|x_{i}) = 1 - S(x_{i}\beta) $

The sigmoid function is what transforms the linear regression model into this logistic classification model as it forces the values $x_{i}\beta$ to be between 0 and 1 (Bernoulli random variable).

$lim _{t\rightarrow-\infty}S(t) = 0$ \
$lim _{t\rightarrow\infty}S(t) = 1$

The sigmoid function takes the following equation:

$S(t) = \frac{1}{1 + exp(-t)} $

and can be visualised as:

Logistic Activation Function.PNG

Once the training set of each dataset is ran through the logistic regression algorithm, the model can then take in new data i.e., the test set. If the probability of the predicted value of $y_{i}$ given the feature values $x_{i}$ is greater than the threshold $0.5$, it will be assigned to one class. If $S(t)$ is less than $0.5$, the model determines that it is part of the other class. This process is iterated through each class until every value of the test set is assigned its predicted class based off the logistic algorithm.

## 2.1 Scatter Plot

```python
#Importing make_classification to create synthetic data
from sklearn.datasets import make_classification

#Randomly generating a 2d, 4 classes dataset with 80 samples per class
X1, y1 = make_classification(n_classes=4,
                             n_features=2,
                             n_samples=320,
                             n_redundant=0,
                             n_informative=2,
                             n_clusters_per_class=1,
                             random_state=1)

#Creating a data frame
df = DataFrame(dict(x=X1[:,0], y=X1[:,1], label=y1))

#Making Four classes
colors = {0:"blue", 1:"orange", 2:"green", 3:"red"}

#Making the Figure
fig, ax = plt.subplots()
grouped = df.groupby("label")

#Creating the scatter plot
for key, group in grouped:
  group.plot(ax=ax, kind="scatter", x="x", y="y", label=key, color=colors[key])

#Showing the plot
plt.show()

#Counting the data elements in each class to ensure 75 in each
df['label'].value_counts()
```

```python
df.head()
```

## 2.2 Training & Test Data

```python
#Importing train_test_split to divide data
from sklearn.model_selection import train_test_split

#Training: dataset split into 75% training and 25% testing
X_train,X_test,y_train,y_test = train_test_split(X1, y1, test_size=0.25)
```

## 2.3 Logistic Regression

```python
#Training the logistic regression model using the training set
LR_model = LogisticRegression()

#Fitting a Logistic Regression model to data
LR_model.fit(X_train, y_train)
```

##2.4 Decision Boundary

```python
#Plotting a decision boundary on the training set
plot_decision_regions(X_train, y_train, clf=LR_model, legend=2)
```

##2.5 Creating Predictions

```python
#Predicting the class label using the test set
y_pred = LR_model.predict(X_test)

#Showing the X1 & y1 inputs and the corresponsing predicted outputs
for i in range(len(X_test)):
  print("X{0} = {1}, Class Predicted = {2}".format(i, X_test[i], y_pred[i]))
```

```python
#Creating a new dataframe
df_new = DataFrame(dict(x=X_test[:,0], y=X_test[:,1], label=y_pred))
```

```python
#Identifying four classes
colors2 = {0:"blue", 1:"orange", 2:"green", 3:"red"}

#Making Figure 2
fig2, ax2 = plt.subplots()

#Creating new data
grouped2 = df_new.groupby("label")

#Creating a Scatter Plot
for key2, group2 in grouped2:
  group2.plot(ax=ax2, kind="scatter", x="x", y="y", label=key2, color=colors2[key2])

#Showing the plot
plt.show()
```

# 2.6 Plotting Predictions

```python
#Replotting the decision boundary on test from the Logistic Regression model
plot_decision_regions(X_test, y_test, clf=LR_model, legend=2)
```

# 2.7 Accuracy Rate

```python
#Calculating the Accuracy Rate
def acc_test(y_pred, y_test):
  correct=0
  for i in range(len(y_test)):
    if y_pred[i] == y_test[i]:
      correct += 1
    else:
      correct += 0
  return correct
acc_test(y_pred, y_test)/len(y_test)
```
